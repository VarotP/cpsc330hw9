### Caveats
There are always some caveats when applying machine learning to real world problems. <br> 
The first caveat is in the problem statement itself, we are using reviews_per_month as a proxy for popularity, which is not necessarily true in the real world. For example when a new restaurant opens there are a lot of inital reviews due to the hype surrounding the opening, and there could be some cases similar to this for Airbnb listings as well. <br> 
Second caveat is that during the feature engineering step, I used the latest date found in the dataset as the date of data collection, which I could not confirm on Kaggle or elsewhere that it is the actual date of data collection, and in the feature importances step I discovered that it was the most important feature in terms of predicting the popularity. <br>
Lastly due to the data collection method, the dataset fails to capture the temporal aspect of popularity, as Airbnb listings and hotels get more customers during high seasons and some even close during low season, and the final prediction is a prediction of the popularity of the listing at the date of data collection, which is in 2019.
